```python
from pyecharts.charts import Pie
import pyecharts.options as opts

#This is the Dictionary from Incopat
legal_dict = {'Microsoft Technology Licensing, Llc':{'Valid':1074, 'Unconfirmed': 319, 'Invalid':153},\
              'Amazon Technologies, Inc.':{'Valid':156, 'Unconfirmed': 41, 'Invalid':4},\
              'Google Inc.':{'Valid':339, 'Unconfirmed': 32, 'Invalid':30, 'Lapse after granting':2}}
```

```python
def legal_status(companyname):
    '''
    Input: (str) the company name that we wish to check
    Output: (html) the pie chart
    '''
    M = legal_dict['{}'.format(companyname)]
    num = []
    lab = []
    for k,v in M.items():
        num.append(v)
        lab.append(k)

    Legalstatus = (Pie(init_opts=opts.InitOpts(width='600px', height='400px')) #Assigning canvas size
        .add(series_name='', 
             radius=['40%', '75%'], #Setting the inner and outer ring size
             data_pair=[(j, i) for i, j in zip(num, lab)],
             center=["50%", "50%"]) #loop through data
        .set_global_opts(title_opts=opts.TitleOpts(title="Legal Status of Patent",subtitle="{}".format(companyname), pos_bottom = '30')) #Add the Titles
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {d}%")) #Add the labels
    ).render_notebook()
    
    return Legalstatus
```

```python
legal_status('Microsoft Technology Licensing, Llc')
```

```python
legal_status('Amazon Technologies, Inc.')
```

```python
legal_status('Google Inc.')
```
