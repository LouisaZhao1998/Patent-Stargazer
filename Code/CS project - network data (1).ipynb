# 1. clean network data

```python
def network_data(cited,citedby,company):
    df1 = pd.read_csv(cited, header = None,names = ['target'],delimiter = "\t")
    df2 = pd.read_csv(cited, header = None,names = ['source'],delimiter = "\t")
    df1['source'] = company
    df2['target'] = company
    order = ['source','target']
    merge = df1[order].append(df2)
    return merge
```

```python
#centrality measure
degree_cen = nx.degree_centrality(g)
dgc_lst = sorted(degree_cen.items(), reverse = True, key = lambda x:x[1])[:10]
dgc_lst
```

```python
#eigenvector centrality
eigen_cen = nx.closeness_centrality(g)
en_lst = sorted(eigen_cen.items(), reverse = True, key = lambda x:x[1])[:10]
en_lst
```

```python
#Betweenness
between_cen = nx.betweenness_centrality(g)
between_cen_lst = sorted(between_cen.items(), reverse = True, key = lambda x:x[1])[:10]
between_cen_lst
```

```python
pagerank_g = nx.pagerank(g)
pagerank_lst = sorted(pagerank_g.items(), reverse = True, key = lambda x:x[1])[:10]
pagerank_lst
```

```python
pr_df = pd.DataFrame(pagerank_lst,columns=['company','pagerank score'])
btw_df = pd.DataFrame(between_cen_lst,columns=['company','betweeness centrality'])
en_df = pd.DataFrame(en_lst,columns=['company','eigenvector centrality'])
dgc_df = pd.DataFrame(dgc_lst,columns = ['company','degree centrality'])
```

```python
frames = [pr_df,btw_df,en_df,dgc_df]
result = pd.concat(frames,axis=1)
```
