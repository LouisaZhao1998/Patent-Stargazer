```python
import tkinter as tk
from tkinter import *

# pip install pillow
from PIL import Image, ImageTk
```

```python
top = tk.Tk()
top.wm_title("Patent Stargazer")
top.geometry("500x200")
label = Label(top, text="Patent Stargazer", font=("Arial Bold", 30))
label.pack()
```

```python
def microhi():
    class Window(Frame):
        def __init__(self, master=None):
            Frame.__init__(self, master)
            self.master = master
            self.pack(fill=BOTH, expand=1)

            loadlegalstatus = Image.open("mls.png")
            loadcpc = Image.open("Microsoft_cpc.png")
            loadcity = Image.open("mcity.png")
            loadyear = Image.open("mtime.png")
            loadnlp = Image.open("mnlpn.png")
            loadnetwork = Image.open("mnwn.png")
            
            #time appears at the 1st place
            renderd = ImageTk.PhotoImage(loadyear)
            imgb = Label(self, image=renderd)
            imgb.image = renderd
            imgb.place(x=0, y=0)            
                  
            #legal status 2nd place
            rendera = ImageTk.PhotoImage(loadlegalstatus)
            imga = Label(self, image=rendera)
            imga.image = rendera
            imga.place(x=400, y=0)

            #cpc 3rd/4th place
            renderb = ImageTk.PhotoImage(loadcpc)
            imgb = Label(self, image=renderb)
            imgb.image = renderb
            imgb.place(x=690, y=330)
            
            #city 4th/3rd place
            renderc = ImageTk.PhotoImage(loadcity)
            imgc = Label(self, image=renderc)
            imgc.image = renderc
            imgc.place(x=890, y=0)
            
            rendere = ImageTk.PhotoImage(loadnlp)
            imgb = Label(self, image=rendere)
            imgb.image = rendere
            imgb.place(x=0, y=330)
            
            renderf = ImageTk.PhotoImage(loadnetwork)
            imgb = Label(self, image=renderf)
            imgb.image = renderf
            imgb.place(x=1010, y=330)

    root = Toplevel()
    app = Window(root)
    root.wm_title("Microsoft")
    root.geometry("2000x1200")
    root.mainloop()    
```
